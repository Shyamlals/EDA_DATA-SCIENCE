# Linear regression

```python
# Create and fit the linear regression model
from sklearn.linear_model import LinearRegression
# Prepare the independent variable (fund managers)
X = pd.get_dummies(df['fund_manager'], drop_first=True)

# Prepare the dependent variable (returns)
y = df['returns_1yr']
model = LinearRegression()
model.fit(X, y)
```

If a fund manager has a positive coefficient, it suggests that mutual funds managed by that manager tend to have higher returns.

If a fund manager has a negative coefficient, it suggests that mutual funds managed by that manager tend to have lower returns.

```python
# Get the coefficients and their corresponding fund managers
coefficients = pd.Series(model.coef_, index=X.columns)

# Separate positive and negative coefficients
positive_coeffs = coefficients[coefficients > 1.5]
negative_coeffs = coefficients[coefficients < -1.0]
```

```python
# Plotting the coefficients for fund managers with positive coefficients
plt.figure(figsize=(12, 6))
positive_coeffs.plot(kind='bar', color='blue')
plt.xlabel('Fund Manager')
plt.ylabel('Coefficient')
plt.title('Coefficients for Fund Managers with Positive Coefficients')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Filter fund managers with positive coefficients greater than 1.5
positive_managers = coefficients[coefficients > 1.5].index.tolist()

# Filter fund managers with negative coefficients less than -1
negative_managers = coefficients[coefficients < -1].index.tolist()

print("Fund Managers with Positive Coefficients (> 1.5):")
print(positive_managers)

print("\nFund Managers with Negative Coefficients (< -1):")
print(negative_managers)
```

# Hypothesis

Null hypothesis (H0): The average expense ratio of mutual funds in the "Equity" category are equal to the average expense ratio of mutual funds in the "Debt" category.

Alternative hypothesis (H1): The average expense ratio of mutual funds in the "Equity" category are significantly different from the average expense ratio of mutual funds in the "Debt" category.

## T test

```python
from sklearn.model_selection import StratifiedShuffleSplit
from scipy import stats
```

```python
#random sampling
np.random.seed(42)
equity_sample = df[df['category'] == 'Equity'].sample(n=30, random_state=42)
debt_sample = df[df['category'] == 'Debt'].sample(n=30, random_state=42)



```

```python
equity_expense_ratio = equity_sample['expense_ratio']
debt_expense_ratio = debt_sample['expense_ratio']

# Perform the t-test assuming equal variances
t_statistic, p_value = stats.ttest_ind(equity_expense_ratio, debt_expense_ratio, equal_var=True)
```

```python
print("T-Statistic:", t_statistic)
print("P-value:", p_value)
```

In this case, the z-statistic is negative, indicating that the average expense ratio of the "Equity" category are lower than the average expense ratio of the "Debt" category.

 The extremely low p-value (close to zero) suggests strong evidence against the null hypothesis.


The evidence suggests that the average expense ratio of the "Equity" category are significantly lower than the average expense ratio of the "Debt" category.
